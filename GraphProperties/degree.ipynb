```python
import igraph as ig
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import powerlaw

from tqdm import trange, tqdm

from collections import defaultdict, Counter
from itertools import combinations

from load_graph import *
```

```python
## DBLP graph
real, _ = load_snap("data/com-dblp.ungraph.txt", "data/com-dblp.all.cmty.txt")
abcdoo_d2, _ = load_abcdoo("data/abcdoo_dblp_d2_edge.dat", "data/abcdoo_dblp_d2_com.dat")

fig, ax = plt.subplots()
options = [real, abcdoo_d2]
names = ["DBLP", r"ABCD+O$^2$"]
for g, name in zip(options, names):
    degrees = g.degree()
    X = powerlaw.Fit(degrees, xmin=np.min(degrees))
    X.plot_ccdf(ax=ax, linewidth=3, original_data=False, linestyle=":", label=name)

ax.set_title("DBLP Degree Distribution", fontsize=16)
ax.legend()
ax.set_xlabel("Degree", fontsize=15)
ax.set_ylabel("Inverse CDF", fontsize=15)
ax.tick_params(labelsize=12)
fig.set_dpi(140)
```
