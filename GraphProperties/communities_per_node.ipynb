```python
import matplotlib.pyplot as plt
import powerlaw

from load_graph import *
```

```python
## DBLP graph
g, coms = load_snap("data/com-dblp.ungraph.txt", "data/com-dblp.all.cmty.txt")
real = g.vs["comms"]
_, abcdoo_d2 = load_coms("data/abcdoo_dblp_d2_com.dat", True)
_, abcdoo_d5 = load_coms("data/abcdoo_dblp_d8_com.dat", True)
_, abcdoo_d10 = load_coms("data/abcdoo_dblp_d64_com.dat", True)
_, cbk = load_coms("data/cbk_dblp.dat", True)

options = [real, abcdoo_d2, abcdoo_d5, abcdoo_d10, cbk]
names = ["DBLP", "d=2", "d=8", "d=64", "CBK"]
fig, ax = plt.subplots(dpi=400)
for cs, name in zip(options, names):
    X = powerlaw.Fit([len(x) for x in cs], xmin=1)  # actual fit does not matter
    X.plot_ccdf(ax=ax, linewidth=3, original_data=False, linestyle=":", label=name)

ax.set_title("DBLP-Like Communities Per Node Distribution", fontsize=15) # At 16 title overflows figure
ax.legend(fontsize=13)
ax.set_xlabel("Number of Communities", fontsize=15)
ax.set_ylabel("Inverse CDF", fontsize=15)
ax.tick_params(labelsize=12)
```
